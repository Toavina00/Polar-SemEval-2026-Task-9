```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataframes = {}

for sub in [1, 2, 3]:
    print(f" Subtask {sub} ".center(20, "#"))
    dataframes[f"subtask{sub}"] = {}
    for lang in ["eng", "hau", "swa"]:
        df = pd.read_csv(f"subtask{sub}/train/{lang}.csv")

        print("Language:", lang)
        print("Number of observations:", len(df))
        
        dataframes[f"subtask{sub}"][lang] = df
```

```python
# Dataset size distribution
fig, ax = plt.subplots(1, 1, figsize=(8, 6))
langs = ["eng", "hau", "swa"]
sizes = [len(dataframes["subtask1"][lang]) for lang in langs]

ax.bar(langs, sizes, color=['#1f77b4', '#ff7f0e', '#2ca02c'])
ax.set_ylabel('Number of Observations')
ax.set_title('Dataset Size by Language')
ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig("lang_distribution.pdf", format='pdf')
plt.show()

# Subtask 1: Label imbalance across languages
sub1_data = dataframes["subtask1"]
labels = [col for col in sub1_data["eng"].columns.drop(["id", "text"])]

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

for lang_idx, lang in enumerate(["eng", "hau", "swa"]):
    ax = axes[lang_idx]
    df = sub1_data[lang]
    
    x = np.arange(len(labels))
    width = 0.35
    
    counts_0 = [len(df) - df[label].sum() for label in labels]
    counts_1 = [df[label].sum() for label in labels]
    
    ax.bar(x - width/2, counts_0, width, label='0', alpha=0.8)
    ax.bar(x + width/2, counts_1, width, label='1', alpha=0.8)
    
    ax.set_xlabel('Labels')
    ax.set_ylabel('Count')
    ax.set_title(f'Subtask 1: {lang.upper()}')
    ax.set_xticks(x)
    ax.set_xticklabels(labels, rotation=45, ha='right')
    ax.legend()
    ax.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig("s1_label_distribution.pdf", format='pdf')
plt.show()

# Subtask 2: Label imbalance per language
sub2_data = dataframes["subtask2"]
labels = [col for col in sub2_data["eng"].columns.drop(["id", "text"])]

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

for lang_idx, lang in enumerate(["eng", "hau", "swa"]):
    ax = axes[lang_idx]
    df = sub2_data[lang]
    
    x = np.arange(len(labels))
    width = 0.35
    
    counts_0 = [len(df) - df[label].sum() for label in labels]
    counts_1 = [df[label].sum() for label in labels]
    
    ax.bar(x - width/2, counts_0, width, label='0', alpha=0.8)
    ax.bar(x + width/2, counts_1, width, label='1', alpha=0.8)
    
    ax.set_xlabel('Labels')
    ax.set_ylabel('Count')
    ax.set_title(f'Subtask 2: {lang.upper()}')
    ax.set_xticks(x)
    ax.set_xticklabels(labels, rotation=45, ha='right')
    ax.legend()
    ax.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig("s2_label_distribution.pdf", format='pdf')
plt.show()

# Subtask 3: Label imbalance per language
sub3_data = dataframes["subtask3"]
labels = [col for col in sub3_data["eng"].columns.drop(["id", "text"])]

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

for lang_idx, lang in enumerate(["eng", "hau", "swa"]):
    ax = axes[lang_idx]
    df = sub3_data[lang]
    
    x = np.arange(len(labels))
    width = 0.35
    
    counts_0 = [len(df) - df[label].sum() for label in labels]
    counts_1 = [df[label].sum() for label in labels]
    
    ax.bar(x - width/2, counts_0, width, label='0', alpha=0.8)
    ax.bar(x + width/2, counts_1, width, label='1', alpha=0.8)
    
    ax.set_xlabel('Labels')
    ax.set_ylabel('Count')
    ax.set_title(f'Subtask 3: {lang.upper()}')
    ax.set_xticks(x)
    ax.set_xticklabels(labels, rotation=45, ha='right')
    ax.legend()
    ax.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig("s3_label_distribution.pdf", format='pdf')
plt.show()
```
